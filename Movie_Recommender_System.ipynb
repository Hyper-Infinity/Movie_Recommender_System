## Creating an importent Tags

```python
# Creating tags from genres column

df.iloc[0, 2]
```

```python
# but this is a string we have to convert this string to python list 
import ast 
ast.literal_eval(df.iloc[0, 2])
```

```python
def tag_creater(obj) :
    obj = ast.literal_eval(obj)
    ans = list()
    for i in obj :
        ans.append(i['name'])
    return ans
```

```python
df['genres'] = df['genres'].apply(tag_creater)
```

```python
# Creating tags from keywords column

df['keywords'][0]
```

```python
tag_creater(df['keywords'][0])
```

```python
df['keywords'] = df['keywords'].apply(tag_creater)
```

```python
df.head()
```

```python
df.iloc[0, 4]
```

```python
# creating tag with top 3 actors in movie from cast column 

df['cast'][0]
```

```python
def top_actor(obj) :
    obj = ast.literal_eval(obj)
    count = 0
    ans = list()
    for i in obj :
        if count == 3 :
            break
        ans.append(i['name'])
        count = count + 1
    return ans
```

```python
top_actor(df['cast'][0])
```

```python
df['cast'] = df['cast'].apply(top_actor)
```

```python
# Creating tags for crew column

ast.literal_eval(df['crew'][0])
```

```python
def tag_crew(obj) :
    obj = ast.literal_eval(obj)
    ans = list()
    
    for i in obj :
        if i['job'] == 'Director' or i['job'] == 'Producer' :
            if (i['name'] not in ans) :
                ans.append(i['name'])
    return ans 
```

```python
tag_crew(df['crew'][0])
```

```python
df['crew'] = df['crew'].apply(tag_crew)
```

```python
# for Overview column 
df['overview'] = df['overview'].apply(lambda x : x.split())
```

```python
df.head()
```

```python
# Removing spaces between tags to avoid confusion between  Sam Worthington and Sam Mendes type fo words

for i in df.columns :
    if i != 'movie_id' and i != 'title' :
        df[i] = df[i].apply(lambda x : [j.replace(" ", "") for j in x])
```

```python
df.head(2)
```

```python
df['Tags'] = df['genres'] + df['keywords'] + df['overview'] + df['cast'] + df['crew']
df = df[['movie_id', 'title', 'Tags']]
```

```python
df.head(2)
```

```python
df['Tags'] = df['Tags'].apply(lambda x : " ".join(x))
```

```python
df.head(2)
```

```python
df['Tags'] = df['Tags'].apply(lambda x : x.lower())
```

```python
# To find Similarities between movies we have to convert each movie into kD vector 

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 5000, stop_words = 'english')
movie_vectors = cv.fit_transform(df['Tags']).toarray()
```

```python
movie_vectors.shape
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_row', None)
pd.DataFrame(cv.get_feature_names_out()).tail(50)
# Problem here is in our data words like loving, loves, loved are different so we have to apply Steming
```

```python
!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text) :
    y = list()
    for i in text.split() :
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem('Loving Samarth went actors words')
```

```python
df['Tags'] = df['Tags'].apply(stem)
```

```python
movie_vectors = cv.fit_transform(df['Tags']).toarray()
```

```python
pd.DataFrame(cv.get_feature_names_out()).tail(50)
```

```python
movie_vectors
```

```python
# As we are working with 5000 dimensional data, euclidean distance is not a Reliable measure so we have to use cosine distance (theta between two vectors)
# distance and similarities are Inversely proportional

from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(movie_vectors)
```

```python
similarity.shape
```

```python
df[df['title'] == 'Batman Begins'].index[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x : x[1])[1 : 6]
```

```python
def recommend(movie) :
    movie_index = df[df['title'] == movie].index[0]
    dist = similarity[movie_index]
    recommend_movie_list = sorted(list(enumerate(dist)), reverse = True, key = lambda x : x[1])[1 : 6]
    
    for i in recommend_movie_list :
        print(df.loc[i[0], 'title'])
```

```python
recommend('Avatar')
```
